```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sbn

from plotter import MixturePlotter
from mixture_distribution import CombinedGaussians
```

# Single Gaussian

```python
sample_size = 10000
mn, mx = 0, 250
```

```python
cg = CombinedGaussians(1, sample_size, lower_bound=mn, upper_bound=mx)#, params=[[240], [255]])
```

```python
MixturePlotter.visualise_sample(cg, params_from_sample=False, visualise_var=True)
```

```python
MixturePlotter.visualise_pmf(cg, figsize=(18,10), visualise_var=True, params_from_sample=False)
```
