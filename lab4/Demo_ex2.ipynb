# Posterior-based Number Game

```python
concepts = gen_base_concepts()
Concept.normalize_priors(concepts)
```

```python
def posteriori_based_nb_game(data, hypotheses):
    results = posteriors(data, hypotheses)
    return OrderedDict(sorted(results.items(), key=lambda t: -t[1]))
```
