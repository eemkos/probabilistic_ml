```python
import numpy as np
from collections import OrderedDict
import math
from estimates_calculations import Concept
from estimates_calculations import likelihood
```

```python
def gen_base_concepts():
    concepts = []
    concepts.append(Concept('odd', lambda x: x%2==1))
    concepts.append(Concept('even', lambda x: x%2==0))
    def checker_gen(v):
        return lambda x: x%v==0
    for i in range(3, 501):
        concepts.append(Concept('multiples_of_%d'%i, checker_gen(i)))
    concepts.append(Concept('pow_of_2', lambda x: 2 ** int(math.log(x,2)) == x))
    return concepts
```

```python
base_concepts = gen_base_concepts()
```

# Likelihood-based Number Game

```python
def likelihood_based_nb_game(data, concepts):
    results = dict()
    for c in concepts:
        results[c.name] = likelihood(data, c)
    results = likelihoods(data, concepts)
    return OrderedDict(sorted(results.items(), key=lambda t: -t[1]))
```

```python
input_numbers = [40,80,160]#np.random.randint(1, 501, 3)

print(input_numbers)
results = likelihood_based_nb_game(input_numbers, base_concepts)
results
```
